### LinearRegression

```python
# boston dataset
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python
boston = datasets.load_boston()
X_train = boston['data']
y_train = boston['target']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_train)
fig, ax = plt.subplots()
sns.scatterplot(y_train, y_pred, color='red',)
ax.set_xlabel('y_train', size=16)
ax.set_ylabel('y_pred', size=16)
ax.set_title('y_train vs y_pred')
sns.despine()
```

```python
predictors = boston.feature_names
print(predictors)
beta_hats = model.coef_
print('\n'.join([f'{predictors[i]}: {round(beta_hats[i], 3)}' for i in range(13)]))
```
