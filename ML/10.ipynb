```python
import tensorflow as tf
from sklearn import datasets
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
boston = datasets.load_boston()
X_boston, y_boston = boston['data'], boston['target']
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=16, activation='relu', input_shape=(X_boston.shape[1],)))
model.add(tf.keras.layers.Dense(units=1, activation='linear'))
model.compile(optimizer='adam', loss='mse')
print(model.summary())
model.fit(X_boston, y_boston, epochs=100, batch_size=1, validation_split=0.2)
```

```python
inputs = tf.keras.Input(shape=(X_boston.shape[1],))
hidden = tf.keras.layers.Dense(units=16, activation='relu')(inputs)
outputs = tf.keras.layers.Dense(units=1, activation='linear')(hidden)
model = tf.keras.Model(inputs=inputs, outputs=outputs)
model.compile(optimizer='adam', loss='mse')
print(model.summary())
model.fit(X_boston, y_boston, epochs=100, batch_size=1, validation_split=0.2)
```

```python
predictions = model.predict(X_boston)[:, 0]
fig, ax = plt.subplots()
sns.scatterplot(y_boston, predictions)
ax.set(xlabel = r"$y$", ylabel = r"$\predictions$", title = r"$y$ vs. $\hat{y}$")
sns.despine()
```
